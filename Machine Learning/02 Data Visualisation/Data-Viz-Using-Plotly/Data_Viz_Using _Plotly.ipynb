```
import plotly.express as px
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(uploaded['DA2_Dataset1.csv']))
fig = px.bar(df, x=' Grd_yr')
fig.show()
```

```
import pandas as pd
import plotly.express as px

df = pd.read_csv(io.BytesIO(uploaded['DA2_Dataset1.csv']))

fig = px.line(df, x = ' Grd_yr', y = ' Grd_yr', title='Line Chart')
fig.show()
```

```
import pandas as pd
import plotly.express as px
df = pd.read_csv(io.BytesIO(uploaded['DA2_Dataset1.csv']))
fig = px.histogram(df, x=" Grd_yr", color=" gndr")
fig.show()
```

```
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
from matplotlib import style
# setting style for graphs
style.use('ggplot')
plt.rcParams['figure.figsize'] = (20,10)
df = pd.read_csv(io.BytesIO(uploaded['DA2_Dataset1.csv']))
result = df.pivot(columns=' pri_spec', values=' Grd_yr')
sb.heatmap(result, annot=False, fmt="g", cmap='viridis')
plt.show()
```

```
import plotly.express as px
import pandas as pd
df = pd.read_csv(io.BytesIO(uploaded['DA2_Dataset1.csv']))
fig = px.scatter(df, x=" pri_spec", y=" st", color=" st")
fig.show()
```
